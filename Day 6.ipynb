```python
import pandas as pd
import numpy as np

# Creating DataFrames
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'Salary': [50000, 60000, 70000, 80000]
}
df = pd.DataFrame(data)
print(df)

# Reading from CSV (comment corrected)
# df = pd.read_csv('data.csv')  # Make sure the quotes are balanced

# Basic operations
print(df.head(2))             # First 2 rows
print(df.describe())          # Summary statistics
print(df['Age'].mean())       # Average age

# Data selection
print(df[df['Age'] > 30])                         # Filtering
print(df.loc[1:2, ['Name', 'Salary']])            # Selection by label

# Handling missing data
df_with_nan = pd.DataFrame({
    'A': [1, 2, np.nan, 4],
    'B': [5, np.nan, np.nan, 8]  # Fixed square brackets and values
})
print(df_with_nan.fillna(value=0))                # Replace NaNs with 0
print(df_with_nan.dropna())                       # Drop rows with NaNs

# Grouping
df['Department'] = ['HR', 'Tech', 'Tech', 'HR']
print(df.groupby('Department').mean())            # Average by department

# Merging data
df2 = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Eve'],
    'Bonus': [2000, 1500, 3000]  # Fixed brackets
})
print(pd.merge(df, df2, on='Name', how='left'))   # Left join

# Time series functionality
date_rng = pd.date_range(start='1/1/2023', end='1/10/2023', freq='D')
ts = pd.DataFrame(date_rng, columns=['date'])
ts['value'] = np.random.randint(0, 100, size=(len(date_rng)))
ts.set_index('date', inplace=True)                # Fixed method syntax
print(ts.resample('3D').mean())                   # Resample to 3-day periods
```

```python
import pandas as pd
import numpy as np

# Creating DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'Salary': [50000, 60000, 70000, 80000]
}
df = pd.DataFrame(data)
print("Original DataFrame:")
print(df)

# Reading from CSV (if needed)
# df = pd.read_csv('data.csv')  # Ensure correct filepath and quotes

# Basic operations
print("\nFirst 2 rows:")
print(df.head(2))

print("\nSummary statistics:")
print(df.describe())

print("\nAverage Age:")
print(df['Age'].mean())

# Data selection
print("\nPeople older than 30:")
print(df[df['Age'] > 30])

print("\nName and Salary of 2nd and 3rd entries:")
print(df.loc[1:2, ['Name', 'Salary']])

# Handling missing data
df_with_nan = pd.DataFrame({
    'A': [1, 2, np.nan, 4],
    'B': [5, np.nan, np.nan, 8]
})
print("\nMissing values replaced with 0:")
print(df_with_nan.fillna(value=0))

print("\nRows with no missing values:")
print(df_with_nan.dropna())

# Grouping
df['Department'] = ['HR', 'Tech', 'Tech', 'HR']
print("\nAverage Age and Salary by Department:")
print(df.groupby('Department')[['Age', 'Salary']].mean())

# Merging data
df2 = pd.DataFrame({
    'Name': ['Alice', 'Bob', 'Eve'],
    'Bonus': [2000, 1500, 3000]
})
print("\nMerged DataFrame (Left Join):")
print(pd.merge(df, df2, on='Name', how='left'))

# Time series functionality
date_rng = pd.date_range(start='1/1/2023', end='1/10/2023', freq='D')
ts = pd.DataFrame(date_rng, columns=['date'])
ts['value'] = np.random.randint(0, 100, size=(len(date_rng)))
ts.set_index('date', inplace=True)

print("\nResampled Time Series (3-day average):")
print(ts.resample('3D').mean())
```
